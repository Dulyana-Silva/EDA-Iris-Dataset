## Handling Outliers
Outlier is a data point that deviates significantly from the rest.

```python
#Detecting outliers
sns.boxplot(data= df)
plt.show()
```

```python
#There are 4 outliers in Sepal width column ---> values below 2 and aove 4
sns.boxplot(x='sepal_width', data= df)
```

## Removing Outliers

We'll detect the outliers using IQR and then remove them from the dataset.

```python
from numpy._core.defchararray import lower
#IQR
Q1= np.percentile(df['sepal_width'], 25, interpolation='midpoint')

Q3= np.percentile(df['sepal_width'], 75, interpolation='midpoint')

IQR= Q3-Q1

print("Old shape= ", df.shape)

#Upper bound
Upper = np.where(df['sepal_width'] >= (Q3+1.5*IQR))

#Lower bound
lower = np.where(df['sepal_width'] <= (Q1-1.5*IQR))

#Removing the Outliers
df.drop(Upper[0], inplace= True)
df.drop(lower[0], inplace= True)

print("New shape= ", df.shape)

sns.boxplot(x= 'sepal_width', data = df)
```
